```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import pearson3
import random
%matplotlib inline
```

```python
skew = 0.1
rv=pearson3.rvs(skew=0.1, size=1000)
x=np.linspace(-3,3,100)
```

```python
mean, var, skew, kurt = pearson3.stats(skew, moments='mvsk')
print(mean,var,skew,kurt)
```

```python
plt.hist(rv,bins=50,normed=True)
plt.plot(x, pearson3.pdf(x, skew),'r', lw=5, alpha=0.6, label='pearson3 pdf')
plt.xlabel("rand. values")
plt.ylabel("Pdf")
```

```python
#суть этого быдлокода очень проста. Во внутреннем цикле стакаются n случайных чиселок, которые являются индексами
#чиселок в выборе. Внешний цикл отсчитывает количество выборок.
#Кстати, функция распределения-распределение Пирсона третьего типа(хотя кому это надо)
#дисперсия посчитана в третьем блоке, где считаются первые несоклько моментов
n=3
a1=np.array([])
x1=np.array([])
random.seed(42)
for i in range(1000):
    for j in range(n):
        x1=np.hstack((x1,np.array([rv[random.randint(0,999)]])))
    x1=np.sum(x1)/n
    a1=np.hstack((a1,x1))
plt.hist(a1,bins=50,normed=True)
norm_rv = sts.norm(0, var/(n**0.5))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
```
